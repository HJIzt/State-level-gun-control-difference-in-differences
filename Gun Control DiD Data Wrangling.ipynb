# 1. Crime Data

```python
sheet_url_crime = "https://docs.google.com/spreadsheets/d/1Q_mmmQcS6p7HehW5QAHUf9xJY70Qc5NDfIl9obAvM60/edit#gid=941088143"
url_crime = sheet_url_crime.replace("/edit#gid=", "/export?format=csv&gid=")

raw_crime_data = pd.read_csv(url_crime)
raw_crime_data.head()
```

### Dropping irrelevant row: FEDERAL

```python
raw_crime_data = raw_crime_data.set_index("jurisdiction")
```

```python
raw_crime_data = raw_crime_data.drop("FEDERAL", axis = 0)
raw_crime_data["violent_crime_100k"] = 100000*raw_crime_data["violent_crime_total"] / raw_crime_data["state_population"]
raw_crime_data["murder_manslaughter_100k"] = 100000*raw_crime_data["murder_manslaughter"] / raw_crime_data["state_population"]
raw_crime_data["agg_assault_100k"] = 100000*raw_crime_data["agg_assault"] / raw_crime_data["state_population"]
raw_crime_data["property_crime_total_100k"] = 100000*raw_crime_data["property_crime_total"] / raw_crime_data["state_population"]

raw_crime_data.head()
```

```python
raw_crime_data_noindex = raw_crime_data.reset_index()
```

```python
raw_crime_data_noindex
```

```python
crime_df = pd.DataFrame({"year": raw_crime_data_noindex["year"],
                         "state": raw_crime_data_noindex["jurisdiction"],
                         "violent_crime_100k": raw_crime_data_noindex["violent_crime_100k"],
                         "murder_manslaughter_100k": raw_crime_data_noindex["murder_manslaughter_100k"],
                         "agg_assault_100k": raw_crime_data_noindex["agg_assault_100k"],
                         "property_crime_total_100k": raw_crime_data_noindex["property_crime_total_100k"]
})
```

# 4. Joining With `crime_df`

```python
crime_df.head()
```

```python
crime_df["state_year"] = crime_df["state"] + " " + crime_df["year"].astype(str)
```

```python
selected_policies = selected_policies.reset_index()
```

```python
selected_policies["state_year"] = selected_policies["state"].str.upper() + " " + selected_policies["year"].astype(str)
```

```python
selected_policies_model_dataset = selected_policies.merge(crime_df, on = "state_year")
```

```python
selected_policies_model_dataset.head()
```

```python
selected_policies_model_dataset["state"] = selected_policies_model_dataset["state_x"]
selected_policies_model_dataset["year"] = selected_policies_model_dataset["year_x"]
selected_policies_model_dataset = selected_policies_model_dataset.drop(["state_x", "year_x", "state_y", "year_y"], axis = 1)
```

```python
selected_policies_model_dataset = selected_policies_model_dataset.fillna(method = "bfill")
selected_policies_model_dataset.head()
```

```python
selected_policies_model_dataset
```

```python
selected_policies_model_dataset["time"] = selected_policies_model_dataset["year"] - 2000
selected_policies_model_dataset["time_squared"] = selected_policies_model_dataset["time"]**2
```

```python
from IPython.display import HTML
import base64  
import pandas as pd  

def create_download_link( df, title = "Download CSV file", filename = "data.csv"):  
    csv = df.to_csv(index =False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)
```

```python
create_download_link(selected_policies_model_dataset)
```
